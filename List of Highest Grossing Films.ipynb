# Import Libraries

```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
import pandas as pd
```

# Get html text out of URL

```python
url = "https://en.wikipedia.org/wiki/List_of_highest-grossing_films"
r = requests.get(url).text
```

# Extract required table as Dataframe

```python
soup = BeautifulSoup(r,"html.parser")
table = soup.find_all("table")[0]
data = pd.DataFrame(columns = ["Rank","Peak","Title","Gross","Year"])
for i in table.tbody.find_all('tr'):
    col = i.find_all("td")
    if col != []:
        title = i.find('a',href = True).get('title')
        data = data.append({"Rank" : col[0].text.strip(),"Peak" : col[1].text.strip(),"Title" : title,"Gross" : col[2].text.strip(),"Year" : col[3].text.strip()},ignore_index = True)
data
```
